```python
# Install required packages
!pip install gradio python-dotenv pillow requests

import gradio as gr
import base64
import requests
import io
from PIL import Image
import os

# API Key
GROQ_API_KEY = ""
GROQ_API_URL = "https://api.groq.com/openai/v1/chat/completions"

if not GROQ_API_KEY:
    raise ValueError("GROQ_API_KEY is missing. Replace it securely using .env or Colab secrets.")

# Function to process the image and query
def process_image(img, query):
    try:
        # Convert the image to base64
        img_bytes = io.BytesIO()
        img.save(img_bytes, format="JPEG")
        encoded_image = base64.b64encode(img_bytes.getvalue()).decode("utf-8")

        messages = [
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": query},
                    {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{encoded_image}"}}
                ]
            }
        ]

        # API Request
        response = requests.post(
            GROQ_API_URL,
            json={"model": "llama-3.2-90b-vision-preview", "messages": messages, "max_tokens": 1000},
            headers={"Authorization": f"Bearer {GROQ_API_KEY}", "Content-Type": "application/json"},
            timeout=30
        )

        # Process response
        if response.status_code == 200:
            result = response.json()
            return result["choices"][0]["message"]["content"]
        else:
            return f"API Error: {response.status_code} - {response.text}"

    except Exception as e:
        return f"Error: {str(e)}"

# Gradio UI
iface = gr.Interface(
    fn=process_image,
    inputs=[
        gr.Image(type="pil", label="Upload Image"),
        gr.Textbox(label="Enter your query")
    ],
    outputs=gr.Textbox(label="Dr. VisionX Prescription"),
    title="Dr. VisionX",
    description="Upload an image and enter  to get a Prescription  from Dr. VisionX."
)

# Launch the UI
iface.launch()
```
